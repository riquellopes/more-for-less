```python
import pandas as pd
```

```python
import json
with open(r"./itens.json") as itens:
  data = json.load(itens)
```

```python
df = pd.DataFrame(data)
df['miles'] = df['miles'].str.replace('.', '', regex=True).astype(int)
df['price'] = df['price'].str.replace('.', '', regex=True).replace(',', '.', regex=True)
df['qty'] = (df['miles']/df['price'].astype(float).astype(int)).astype(int)
df.sort_values(by=['qty'], ascending=False)
```
